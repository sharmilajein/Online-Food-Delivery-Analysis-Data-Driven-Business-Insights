```python
# fill with specific values
df['Customer_Gender'] = df['Customer_Gender'].fillna('Unknown')
df['City'] = df['City'].fillna('Not specified')
```

```python
# Age with average
df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].mean())
df["Customer_Age"] = df["Customer_Age"].round().astype(int)
# Fill Cuisine Type with most Frequent value
df['Cuisine_Type'] = df['Cuisine_Type'].fillna(df['Cuisine_Type'].mode()[0])
```

```python
df[["Customer_Age"]].head()
```

```python
# Handle high missing values

df['Cancellation_Reason'] = df['Cancellation_Reason'].fillna('Not cancelled')


```

```python
print(df['Cancellation_Reason'].isnull().sum())
```

```python
# Area- Fill with mode
df['Area'] = df['Area'].fillna(df['Area'].mode()[0])
# Distance_km - Fill with median(robust to outliers)
df['Distance_km'] = df['Distance_km'].fillna(df['Distance_km'].median())
#Peak_Hour-fill with mode 
df['Peak_Hour'] = df['Peak_Hour'].fillna(df['Peak_Hour'].mode()[0])
#Delivery_Rating - fill with median or 0
df['Delivery_Rating'] = df['Delivery_Rating'].fillna(df['Delivery_Rating'].median())
# Order_Date', 'Order_Time - Better to drop these rows
df.dropna(subset=['Order_Date', 'Order_Time'], inplace=True)
# Drop rows where essential info is missing
df.dropna(subset=['Final_Amount'], inplace=True)


```

```python

# # Missing Values
df['Delivery_Time_Min'] = df['Delivery_Time_Min'].fillna(df['Delivery_Time_Min'].median())
df['Order_Value'] = df['Order_Value'].fillna(df['Order_Value'].median())
df['Payment_Mode'] = df['Payment_Mode'].fillna(df['Payment_Mode'].mode()[0])
```

```python
print("\nPercentage of missing values:")
print((df.isnull().sum() / len(df)) * 100)
```

```python

for col in ['Delivery_Time_Min', 'Order_Value']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    df[col] = np.clip(df[col], Q1 - 1.5*IQR, Q3 + 1.5*IQR)
```

We cap instead of delete because extreme orders can be real.

```python
print("\nPercentage of missing values:")
print((df.isnull().sum() / len(df)) * 100)
```

```python

# Cancelled orders should not have ratings-Invalid Values

df['Delivery_Rating'] = df['Delivery_Rating'].clip(0, 5)
df['Restaurant_Rating'] = df['Restaurant_Rating'].clip(0, 5)
df = df[df['Profit_Margin'] >= 0]

df.loc[df['Order_Status'] == 'Cancelled',
       ['Delivery_Rating', 'Restaurant_Rating']] = np.nan
```
